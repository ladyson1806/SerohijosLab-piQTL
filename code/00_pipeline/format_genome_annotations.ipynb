```python
from os import path

import gffpandas.gffpandas as gffpd
from IPython.display import display, HTML

root_path = path.abspath(path.join(path.abspath(""), "../.."))
```

```python
XUTs_2011_annotation = gffpd.read_gff3(f'{root_path}/data/genome_annotations/van_dikj_2011/van_Dijk_2011_XUTs_V64.gff3')
```

```python
XUTs_2011_annotation.df['seq_id'] = [ XUTs_2011_annotation.df['seq_id'][idx].replace("chr", 'CHR_') for idx in XUTs_2011_annotation.df.index ]
XUTs_2011_annotation.df['locus_id'] = [ XUTs_2011_annotation.df['attributes'][idx].split(';')[0].replace("ID=", 'XUT_') for idx in XUTs_2011_annotation.df.index ]
XUTs_2011_annotation.df['description'] = [ XUTs_2011_annotation.df['attributes'][idx].split(';')[-1].replace("desc=", '') for idx in XUTs_2011_annotation.df.index ]
```

```python
XUTs_2011_annotation.df.to_csv(f'{root_path}/data/genome_annotations/van_dikj_2011/van_Dijk_2011_XUTs_V64.csv', index=False)
```
